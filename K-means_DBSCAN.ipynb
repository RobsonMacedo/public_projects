```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
from sklearn import datasets
```

```python
# biblioteca para a criação de df para testes
x, y = datasets.make_moons(n_samples=1500, noise=0.09)
```

```python
plt.scatter(x[:, 0], x[:, 1], s =5)
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
previsoes = kmeans.fit_predict(x)
```

```python
cores = np.array(['red', 'blue'])
```

```python
plt.scatter(x[:, 0], x[:,1], s = 5, c= cores[previsoes])
```

```python
hc = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
```

```python
previsoes = hc.fit_predict(x)
```

```python
plt.scatter(x[:, 0], x[:,1], s = 5, c= cores[previsoes])
```

```python
dbscan = DBSCAN(eps = 0.1, min_samples=5)
```

```python
previsoes = dbscan.fit_predict(x)
```

```python
# para este tipo de dados o DBSCAN se adapta muito bem
plt.scatter(x[:, 0], x[:,1], s = 5, c= cores[previsoes])
```
